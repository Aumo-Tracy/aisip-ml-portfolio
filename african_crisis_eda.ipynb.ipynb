# African Crisis Dataset - Exploratory Data Analysis


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('african_crises.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
plt.figure()
df['inflation_annual_cpi'].hist()
plt.title('Inflation Distribution')
plt.show()
```

```python
plt.figure()
plt.scatter(df['year'], df['inflation_annual_cpi'])
plt.show()
```

```python
plt.figure()
df['country'].value_counts().head().plot(kind='bar')
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['inflation_annual_cpi'])
plt.show()
```

```python
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```
